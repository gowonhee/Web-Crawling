# 영화 리뷰 데이터 수집 및 워드클라우드
- 영화 리뷰 데이터는 네이버 블로그에서 수집
- 워드클라우드는 형태소분석기(konlpy) 설치 및 사용

```python
import requests as req
from bs4 import BeautifulSoup as bs

# 사용자(브라우저)가 요청했다는 정보를 서버에서 확인할 수 있게 User-Agent 속성 설정 !
header_options = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36'
}
```

```python
n_search_url = 'https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&mra=bkEw&pkid=68&os=2954622&qvt=0&query=%EB%B2%A0%ED%85%8C%EB%9E%912%20%EB%A6%AC%EB%B7%B0'
res = req.get(n_search_url, headers = header_options)
html = bs(res.text, 'lxml')
```

```python
# 영화 리뷰가 작성된 블로그 제목 링크 수집
href_list = []

for a in html.select('dt > a.review_link'):
    href_list.append(a.get('href'))
```

```python
href_list
```

```python
res1 = req.get(href_list[0])
html1 = bs(res1.text, 'lxml')
```

```python
# iframe의 src속성값 접근 -> 접근한 주소값으로 실제 블로그 주소 생성
src = html1.select_one('#mainFrame').get('src')
blog_url = 'https://blog.naver.com'+src
```

```python
blog_url
```

```python
res2 = req.get(blog_url, headers = header_options)
html2 = bs(res2.text, 'lxml')
```

```python
res2 = req.get(blog_url, headers = header_options)
html2 = bs(res2.text, 'lxml')
```

```python
# 스크립트 태그 삭제
for script in html2.select_one('#content-area').select('script'):
    script.extract()
```

```python
content = html2.select_one('#content-area')
content = content.text.replace('\n','').replace('\t','')
content
```

```python
# 수집된 영화리뷰 데이터를 텍스트 파일로 저장
# open(경로, 권한설정): 파일에 접근한 후 내용작성 또는 읽어올 때 사용하는 함수
# 경로: 파일명 또는 폴더명/파일명으로 접근 또는 저장할 위치 설정
# 권한설정 : 쓰기(w), 읽기(r) 중 용도에 따라 설정
# encoding : 파일로 저장하기 위한 인코딩 설정 (기본값: CP949)
#    - CP949는 특수문자나 이모티콘 변환이 안될 수 있으므로 UTF-8로 지정 (이모티콘이나 특수문자가 섞여있으므로 'utf-8'로 인코딩 필요 !)

f = open('movie_review.txt', 'w', encoding='utf-8')
f.write(content)
f.close()
```

# 실습) 모든 영화리뷰 블로그의 내용을 수집한 후 텍스트파일로 저장하기

```python
for href in href_list:
    # 블로그 페이지 접근
    res1 = req.get(href)
    html1 = bs(res1.text, 'lxml')

    # iframe의 src속성값 추출 및 실제 블로그 주소 생성
    src = html1.select_one('#mainFrame').get('src')
    blog_url = 'https://blog.naver.com'+src

    # 실제 블로그 페이지 접근
    res2 = req.get(blog_url, headers = header_options)
    html2 = bs(res2.text, 'lxml')

    # 
    for script in html2.select_one('#content-area').select('script'):
        script.extract()
        
    content = html2.select_one('#content-area')
    content = content.text.replace('\n','').replace('\t','')
```
