```python
# BeautifulSoup: 웹 문서를 Python 객체로 변환해주는 라이브러리
# -> 원하는 데이터 추출 가능
from bs4 import BeautifulSoup as bs
```

```python
n_weather_url = 'https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EA%B4%91%EC%A3%BC+%EB%82%A0%EC%94%A8&oquery=%EB%82%A0%EC%94%A8&tqi=iY4gDdqVOsosstFmcs0ssssst1N-490916'

res = req.get(n_weather_url)

# bs(변환할 데이터, 변환방식)
# - 변환할 데이터: 웹 문서(res.text)
# - 변환방식: lxml > html.parser > html5lib (속도 측면)
html = bs(res.text, 'lxml')

html
```

```python
# 타입 확인
type(res.text), type(html)
```

```python
# select_one(CSS 선택자): 선택자 조건에 맞는 요소를 반환
# 찾는 순선느 위에서 아래로!
# 선택자를 기반으로 찾기 때문에 깊이 들어가서 접근할때 find보다 더 자유도가 높음
html.select_one('title').text
```

```python
html.select_one('strong')
# 위에서부터 정렬이기 때문에 다른곳에 있는 strong이 출력되었음.
```

CSS 선택자 종류
1. 타입선택자 : 요소의 이름으로 선택 ex) div{}, p{}, ...
2. id 선택자 : id값으로 요소를 선택 ex) `<div id="msg"></div>` -> #msg{}
3. class 선택자 : class값으로 요소를 선택 ex) `<div class='item'></div>` -> .item{}
4. 자식선택자 : '>'로 요소를 선택 ex) 부모요소: div / 자식요소: p -> div > p
5. 자손선택자 : ' '(공백)로 요소를 선택 ex) 부모요소: div / 자손요소: a -> div a
6. 인접형제선택자 : '+'로 요소를 선택 ex) 첫번째요소 : li / 두번째요소 li -> li + li
7. 형제선택자 : '~'로 요소를 선택
    ex) h1 p ul p -> 모든 p요소를 선택 -> h1 ~ p 

```python
html.select_one('div.temperature_text> strong').text
```

```python
newsTit = html.select_one('a.news_tit')

print(newsTit)
print(newsTit.text)

# get(요소의 속성이름): 요소의 속성값을 접근하는 메소드 
print(newsTit.get('href'))
print(newsTit.get('class'))
print(newsTit.get('target'))
print(newsTit.get('title'))
```

```python
# select(css 선택자) : 조건에 맞는 모든 요소들을 리스트로 반환
newsTit_list = html.select('a.news_tit')

for newsTit in newsTit_list:
    print(newsTit.text)

# 접근해야 하는 리스트 수가 많을 경우 
for i in range(len(newsTit_list)):
    print(newsTit_list[i].text)
```

# 실습1) 네이버 뉴스 데이터 수집

```python
import time
import random
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
news_url = 'https://n.news.naver.com/mnews/article/366/0001019501'
res = req.get(news_url)
html = bs(res.text, 'lxml')
```

```python
title = html.select_one('#title_area > span').text
print(title)
```

```python
# Tip) 불필요한 요소를 제거한 후에 요소를 접근!
# extract() : 요소를 제거하는 함수 / output (제거된 요소)를 리턴
html.select_one('span.end_photo_org').extract() 
```

```python
content = html.select_one('#dic_area').text
print(content.strip()) # strip : 양쪽 문자열의 공백을 제거
```

# 실습2) 모든 뉴스 데이터 수집

```python
news_url = 'https://news.naver.com/section/105'
res = req.get(news_url)
html = bs(res.text, 'lxml')
```

```python
news_tits = html.select('a.sa_text_title')
href_list = []

for news in news_tits:
    # print(news.get('href'))
    href_list.append(news.get('href'))
```

```python
href_list
# 여러페이지들을 동시에 수집을 하는 경우, 하나의 서버에 요청을 보내는 경우 딜레이를 걸어주는 것이 좋음 !
```

```python
# 2~5 사이로 딜레이 주는 반복문 구현
for href in href_list:
    res = req.get(href)
    html = bs(res.text, 'lxml')

    title = html.select_one('#title_area > span').text
    print('제목:', title)

    html.select_one('span.end_photo_org').extract() 
    content = html.select_one('#dic_area').text
    print(content.strip()) # strip : 양쪽 문자열의 공백을 제거

    print('-'*50)
    time.sleep(random.randint(2,5))
```
