# 유튜브 데이터 수집 실습1)
- 특정 유튜버의 '동영상' 제목, 주소, 조회수 데이터 수집

```python
import time
import pandas as pd

# 테스트 브라우저를 실행해주는 역할
from selenium import webdriver as wb

# 요소를 접근하는 방식을 제공하는 역할
from selenium.webdriver.common.by import By

# 키보드 입력값을 제공하는 역할
from selenium.webdriver.common.keys import Keys
```

```python
yt_url = 'https://www.youtube.com/@forestsquirrel/videos'
driver = wb.Chrome()
```

```python
driver.get(yt_url)
```

```python
# 페이지 스크롤 내리는 방법 ( 요소를 찾는 기준 : body 태그 )
#  - 브라우저 페이지에서 스크롤 내리는 키보드 값: end, page down, space, arrow down
for i in range(5):
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(1.5)
```

```python
yt_videos = driver.find_elements(By.ID, 'video-title-link')
len(yt_videos)
```

```python
# 제목, 영상주소, 조회수 추출
title = yt_videos[0].get_attribute('title')
href = yt_videos[0].get_attribute('href')
reviews = yt_videos[0].get_attribute('aria-label')

print(title)
print(href)
# print(reviews)
# print(reviews.rfind('조회수 ')+4)
# print(reviews.rfind('회'))

# 문자열 사이에 있는 조회수 추출
start = reviews.rfind('조회수 ')+4
end = reviews.rfind('회')

print(reviews[start:end])
```

```python
# 모든 영상제목, 주소, 조회수를 접근한 후 데이터프레임으로 출력해보기!

yt_list = []
reviews_result = reviews[start:end]

for i in range(len(yt_videos)):
    yt_list.append([yt_videos[i].get_attribute('title'),
                    yt_videos[i].get_attribute('href'), 
                    yt_videos[i].get_attribute('aria-label')[start:end]])
yt_list
```

```python
pd.DataFrame(yt_list, columns=['영상제목','주소', '조회수'])
```
